```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris.data
```

```python
data = pd.DataFrame(iris.data,columns=iris.feature_names)
data.head()
```

```python
iris.target
```

```python
data['target']=iris.target
data
```

```python
data[data.target==0].head()
```

```python
data[data.target==2].head()
```

```python
df0 = data[:50] #setosa
df0
```

```python
df1 = data[50:100] #versicolor
df1
```

```python
df2 = data[100:] #virginica
df2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Sepal Length vs Sepal Width (Versicolor vs Virginica)

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Veriscolor vs Virginica')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], color = 'blue', marker = '+')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'], color = 'orange', marker = '*')
plt.show()
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
x = data.drop(['target'],axis=1)
print(x)
```

```python
y=data['target']
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2, random_state=4)
print(len(x_train))
print(len(x_test))
```

Create KNN (K- Nearest Neighbour Classifier)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```
